**Import libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

## **Homes concat**

**Path declaration**

```python
# Specify the folder path
folder_path = "Data\Hogares"

# Create an empty list to store the DataFrames
dfs = []

# Use glob to create a list of file paths with the specified pattern
file_paths = glob.glob(folder_path + "\CNPV2018_2HOG_A2_*.CSV")
```

**Files append loop**

```python
# Iterate through the file paths and concatenate DataFrames
for file_path in file_paths:
    df = pd.read_csv(file_path)  # Read each CSV file into a DataFrame
    dfs.append(df)  # Append the DataFrame to the list
```

**Main dataframe merge**

```python
# Concatenate all DataFrames in the list using pandas.concat()
hog_df = pd.concat(dfs, ignore_index=True)  # Reset index for clarity
hog_df
```

## **Deceased concat**

**Path declaration**

```python
# Specify the folder path
folder_path = "Data\Fallecidos"

# Create an empty list to store the DataFrames
dfs = []

# Use glob to create a list of file paths with the specified pattern
file_paths = glob.glob(folder_path + "\CNPV2018_3FALL_A2_*.CSV")
```

**Files append loop**

```python
# Iterate through the file paths and concatenate DataFrames
for file_path in file_paths:
    df = pd.read_csv(file_path)  # Read each CSV file into a DataFrame
    dfs.append(df)  # Append the DataFrame to the list
```

**Main dataframe merge**

```python
# Concatenate all DataFrames in the list using pandas.concat()
fall_df = pd.concat(dfs, ignore_index=True)  # Reset index for clarity
fall_df
```

### **Data visualization**

**Age range fix**

```python
# X axis limit
x_limit = (0, 130)

# filtered_data = data[data['x'].between(x_limit[0], x_limit[1])]
fall_df = fall_df[fall_df['Age of deceased'].between(x_limit[0], x_limit[1])]
```

**Age of deceased | Histogram**

```python
# Histogram plot
sns.histplot(data = fall_df, x= 'Age of deceased', bins = 10, hue= 'Sex', stat = 'count', multiple= 'dodge')
```

**Age of deceased boxplot by department**

```python
# Filter selected departments
fall_dep_df = fall_df[fall_df['Department'].isin(['BOGOTA, D.C.', 'ANTIOQUIA', 'CHOCO', 'LA GUAJIRA'])]

# Boxplot setting
sns.boxplot(data = fall_dep_df, y = 'Age of deceased', x= 'Department')
plt.title("Age of deceased  by department")
plt.show()
```
